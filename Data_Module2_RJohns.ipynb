```python

import numpy as np
from scipy import stats



A_data = np.load('A_data.npy')
#print(A_data.itemsize, A_data.ndim, A_data.dtype, A_data.shape,)
#A_data[:,:1]
```

```python
#1: print the mean/median/mode of x and y
print("mean of x, y: ", np.mean(A_data, axis = 0), "\n")
print("median of x, y: ", np.median(A_data, axis = 0), "\n")
print("mode of x, y: ", stats.mstats.mode(A_data, axis = 0))
```

```python
#2: sample variance for X and Y
np.var(A_data, axis = 0)
```

```python
#3: plot histogram for x and y
import matplotlib.pyplot as plt
%matplotlib inline


x = A_data[:,:1]
y =  A_data[:,1:]


plt.hist(x, alpha = 0.7,);
plt.hist(y, alpha = 0.3,);
plt.title(' X and Y Histogram')
plt.xlabel('other numbers')
plt.ylabel('numbers');

```

```python
#4: draw scatterplot of first 3000 values in X and Y

plt.scatter(A_data[:3000,:1],A_data[:3000,1:], s = 5);
```

```python
#5: mean vector and covariance matrix for the joint dist. of x and y

print(np.mean(A_data, axis = 0), "\n")
print(np.cov(A_data.T),)

#print(np.cov(x.T,y.T))


```

```python
#6: calculate the correlation between X and Y
print(np.corrcoef(A_data.T))



```

```python
#7: create categorical variable dX,dY by discretizing X,Y into 3 equal width ranges.

dX = np.histogram(x, bins = 3)
dY = np.histogram(y, bins = 3)

print(dX,'\n',dY)
```
