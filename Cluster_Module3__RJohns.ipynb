```python
import numpy as np
import matplotlib.pyplot as plt
# from matplotlib import style
# style.use("ggplot")

from sklearn.cluster import KMeans
from sklearn import metrics
from mpl_toolkits import mplot3d

cdata1 = np.load('cdata1.npy')
cdata2 = np.load('cdata2.npy')
seed_data1 = np.load('seed_data.npy')
seed_data = seed_data1[:,:7]
c=seed_data1[:,7]
class_label = c

# print(cdata1.itemsize, cdata1.ndim, cdata1.dtype, cdata1.shape,)
# cdata1[:,:]
# print(cdata2.itemsize, cdata2.ndim, cdata2.dtype, cdata2.shape,)
# cdata2[:,]
# print(seed_data.itemsize, seed_data.ndim, seed_data.dtype, seed_data.shape,)
# seed_data[:,]
# class_label.ndim
```

```python
%matplotlib inline

#1: examining scattorplot of 2d data set cdata1, I notice 4 clusters

plt.scatter(cdata1[:,:1],cdata1[:,1:], s = 5);
```

```python
#2,3 K-means algorythm on cdata1 starting with 2 cluster centers, 
#plus a 2d scatterplot of results

kmeans = KMeans(n_clusters=2)
kmeans.fit(cdata1)

y_kmeans = kmeans.predict(cdata1)

plt.scatter(cdata1[:, 0], cdata1[:, 1], c=y_kmeans,s=25,)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=75, alpha = .8);



```

```python
#4 silhoutte score for my 2 clusters

labels = kmeans.labels_
metrics.silhouette_score(cdata1, labels)
```
